### ① 取出今天所有的發文

```python
import requests as rq
from bs4 import BeautifulSoup

url="https://www.ettoday.net/news/news-list.htm"
r=rq.get(url)

soup=BeautifulSoup(r.text,"html5lib")
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd 

date=[]
group=[]
title=[]

for d in soup.find(class_="part_list_2").find_all('h3'):
    date.append(d.find_all('span')[-1].text)
    group.append(d.find_all('em')[-1].text)
    title.append(d.find_all('a')[-1].text)


dict1={'class':group, 'date':date, 'title':title}

select_df=pd.DataFrame(dict1)

#依類別進行分類
DataProcess=select_df.sort_values(by=['class', 'date'], ascending = False)
print(DataProcess)
```

### ③ 哪一個類別的文章最多

```python
#established dictionary

class_dict={}

for d in soup.find(class_='part_list_2').find_all('h3'):
    
    if (d.find_all('em')[-1].text) not in class_dict:
        class_dict[(d.find_all('em')[-1].text)]=1
    else:
        class_dict[(d.find_all('em')[-1].text)]+=1
        
print("此網頁共有{}則新聞".format(sum(list(class_dict.values()))))
print(class_dict)

# >>>"{} {}".format("hello", "world")    # 不设置指定位置，按默认顺序
# 'hello world'
 
# >>> "{0} {1}".format("hello", "world")  # 设置指定位置
# 'hello world'
 
# >>> "{1} {0} {1}".format("hello", "world")  # 设置指定位置
max_key=max(class_dict,key=class_dict.get)
print('\n')

print(max_key,"類別文章最多,有", class_dict[max_key],"筆")
```
