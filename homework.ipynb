### 1. 根據範例使用 selenuim 取出 Hahow 網站中所有課程的資料，並用一個適合的資料結構儲存（Hint: 將所有資料存在一個變數當中取出）

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）


# 取得資料
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'lxml')
soup
```

```python
course = []
    
for d in soup.find_all('h4', class_='title'):
    course.append(d.text)

print(course)
```

```python
browser.close()
```

### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
# 1. 先取出全部有幾頁
    
'''
Your Code
'''
total_page=soup.find_all(class_='rc-pagination-item')[-1].text
print(total_page)
      

```

```python
# 2. 每頁都取出資料放入 courses

courses = []
import time, random

delay_times = [8, 5, 10, 6, 20, 11]  #延遲的秒數

#第一頁資料

for d in soup.find_all('h4'):
    courses.append(d.text)
    
for i in range(int(total_page)-1):
    try:
        next_page = browser.find_element_by_class_name('rc-pagination-next')
        next_page.click()
        time.sleep(random.choice(delay_times))

        html_source = browser.page_source
        soup = BeautifulSoup(html_source, 'html.parser')
        
        for j in soup.find_all('h4'): 
            courses.append(j.text)
    except:
        print('搜尋完畢')
        continue

 
```

```python
# 3. done!
browser.close()
print(courses)
```
